```python
import os
import pandas as pd
from zipfile import ZipFile
import matplotlib.pyplot as plt
from matplotlib.widgets import SpanSelector
import numpy as np

# Pfade zu den ZIP-Ordnern
zip_folder_paths = [
    r'C:\Users\mguen\OneDrive\Desktop\VA Project\Data_Covid.zip',
    r'C:\Users\mguen\OneDrive\Desktop\VA Project\Data_stocks.zip'
]

dataframes = []

for zip_folder_path in zip_folder_paths:
    # ZIP-Datei öffnen
    with ZipFile(zip_folder_path, 'r') as zip_file:
        # Liste der Dateinamen in der ZIP-Datei
        zip_file_names = zip_file.namelist()

        for file_name in zip_file_names:
            # Überprüfen, ob es sich um eine CSV-Datei handelt
            if file_name.endswith(".csv"):
                # Extrahiere die CSV-Datei aus dem ZIP-Archiv
                with zip_file.open(file_name) as csv_file:
                    # CSV-Datei lesen und in ein DataFrame konvertieren
                    df = pd.read_csv(csv_file)
                    # Füge den Dateinamen als Spalte hinzu
                    filename = os.path.splitext(file_name)[0]
                    df['Dateiname'] = filename
                    # Füge das DataFrame zur Liste hinzu
                    dataframes.append(df)

# Concateniere alle DataFrames
final_df = pd.concat(dataframes, ignore_index=True)

# Zeitleiste oben
fig, ax_time = plt.subplots()
ax_time.plot(final_df['Date'].astype(str), final_df['Close'])
ax_time.set_xlabel('Datum')
ax_time.set_ylabel('Kurs')

# Streudiagramm
fig, ax_scatter = plt.subplots()
ax_scatter.scatter(final_df['Volume'], final_df['Close'])
ax_scatter.set_xlabel('Volumen')
ax_scatter.set_ylabel('Kurs')

# Balkendiagramm (Filter)
fig, ax_bar = plt.subplots()
ax_bar.bar(final_df['Date'].astype(str), final_df['Volume'])
ax_bar.set_xlabel('Datum')
ax_bar.set_ylabel('Volumen')

# ...
# (Weitere Diagramme und Interaktionen hier einfügen)
# ...

plt.show()
```
